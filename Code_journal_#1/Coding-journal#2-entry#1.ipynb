In a new cell use the function to set y=exp(x), and then plot x vs. y. Label the x-axis as “Time [milliseconds]” and the y-axis as “Awesomeness”. Save the figure as a PDF.  


```python
x = np.arange(0, 1, 0.01)
y = np.exp(x)

plt.plot(x, label ='Time [milliseconds]', color = 'red')
plt.plot(y, label ='Awesomeness', color = 'b')

plt.xlabel(r'Time [milliseconds]')
plt.ylabel(r'Awesomeness')

plt.savefig('Code-Journal#2-Entry#1.pdf', bbox_inches='tight',)
```
